```python
#identifier="SHG:601360"
#identifier.split(":")[0]
#3mapp[identifier.split(":")[0]]
#df[df['country']=='China']

def map_to_ticker(row_value):
    
    if (row_value['identifier'] is None or row_value['identifier'] in ['-']):
        return '-'
    if ('America' in row_value['country']):
        identifier_parts = row_value['identifier'].split(":")
        if len(identifier_parts) > 1:
            var1 = identifier_parts[1]
            return var1
    if ('Egypt' in row_value['country']):
        return row_value['identifier']
    if ('Emirates' in row_value['country']):
        return row_value['identifier']
    if (":" in row_value['identifier']):
        iden = row_value['identifier'].split(":")[0]
        stx = row_value['identifier'].split(":")[1]
        if ("." in stx):
            stx = stx.replace(".", "-")
            if yahoo_exchange_codes[iden] is not None:
                return stx+"."+yahoo_exchange_codes[iden]
            else:
                return stx
        else:
            if yahoo_exchange_codes[iden] is not None:
                return stx+"."+yahoo_exchange_codes[iden]
    else:
        return row_value['identifier']
    
```

```python
import pandas as pd
import yfinance as yf

def calculate_yearly_returns_volatility(ticker):
    try:
        data = yf.download(ticker, start='2024-01-01', end='2024-12-31')
        if data.empty:
            return None

        # Calculate daily returns
        data['Daily Return'] = data['Close'].pct_change()

        # Calculate yearly return
        yearly_return = (1 + data['Daily Return']).prod() - 1

        # Calculate yearly volatility (annualized)
        yearly_volatility = data['Daily Return'].std() * (252**0.5)

        return (round(float(yearly_return),4), round(float(yearly_volatility),4))
    except Exception as e:
        print(f"An error occurred: {e}")
        return (-99, -99)
```
