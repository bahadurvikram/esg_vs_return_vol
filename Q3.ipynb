```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from scipy import stats
from statsmodels.formula.api import ols

# Set visual style
# plt.style.use('seaborn')
sns.set_palette("husl")
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("data/esg_return_vol_clean_data.csv")
```

```python
def find_extreme_outliers(df, column_name):
    try:
        # Calculate quartiles
        Q1 = df[column_name].quantile(0.25)
        Q3 = df[column_name].quantile(0.75)
        IQR = Q3 - Q1

        # Define bounds for extreme outliers (3 times the IQR)
        lower_bound = Q1 - 3 * IQR
        upper_bound = Q3 + 3 * IQR

        # Identify extreme outliers
        extreme_outliers = (df[column_name] < lower_bound) | (df[column_name] > upper_bound)
        return extreme_outliers

    except TypeError:
        print(f"Column '{column_name}' is not numeric and cannot be used to find outliers using the IQR method.")
        return None
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
        return None
```

```python
df_use=df[~find_extreme_outliers(df, 'return')]
df_use=df_use[~find_extreme_outliers(df_use, 'volatility')]
print('Data is use ',len(df_use))
```

```python
def prepare_market_data(df):
    """Prepare data for market comparison analysis"""
    
    # Create market classification (if not already present)
    if 'market_class' not in df.columns:
        # Example classification - you should replace with your actual classification
        developed_markets = ['United States of America', 'United Kingdom', 'Germany', 'Japan', 'Canada', 'France']
        df['market_class'] = df['country'].apply(
            lambda x: 'Developed' if x in developed_markets else 'Emerging'
        )
    
    # Create risk rating order
    risk_order = ['Negligible Risk', 'Low Risk', 'Medium Risk', 'High Risk', 'Severe Risk']
    df['risk_rating_assessment'] = pd.Categorical(
        df['risk_rating_assessment'],
        categories=risk_order,
        ordered=True
    )
    
    return df
def prepare_volatility_data(df):
    """Clean and prepare data for volatility analysis"""
    # Convert numeric columns
    num_cols = ['volatility', 'risk_rating_score']
    for col in num_cols:
        df[col] = pd.to_numeric(df[col], errors='coerce')
    
    # Clean market classification
    df['market_class'] = df['market_class'].astype(str).str.strip()
    df = df[df['market_class'].isin(['Developed', 'Emerging'])]
    
    return df.dropna(subset=['volatility', 'risk_rating_score', 'market_class'])

def plot_volatility_relationships(df):
    """Create visualizations with properly aligned elements"""
    plt.figure(figsize=(20, 16))
    
    # 1. ESG Score vs Volatility by Market Type
    plt.subplot(2, 2, 1)
    sns.scatterplot(x='risk_rating_score', y='volatility', hue='market_class',
                   data=df, alpha=0.6, palette='viridis', s=100)
    plt.title('ESG Score vs. Volatility by Market Type', pad=20, fontsize=14)
    plt.xlabel('ESG Score (Lower = Better)', fontsize=12)
    plt.ylabel('Annualized Volatility', fontsize=12)
    plt.legend(title='Market Type', bbox_to_anchor=(1.02, 1), loc='upper left')
    
    # 2. Boxplot with proper spacing
    plt.subplot(2, 2, 2)
    ax = sns.boxplot(x='risk_rating_assessment', y='volatility', hue='market_class',
                    data=df, showfliers=False, 
                    width=0.6,  # Control box width
                    gap=0.2,    # Space between hue groups
                    palette='viridis')
    
    # Adjust x-tick positions to prevent overlap
    for i, tick in enumerate(ax.get_xticklabels()):
        tick.set_rotation(45)
        tick.set_ha('right')
    
    plt.title('Volatility Distribution by ESG Risk Category', pad=20, fontsize=14)
    plt.xlabel('ESG Risk Category', fontsize=12)
    plt.ylabel('Volatility', fontsize=12)
    plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left')
    
    # Add space between boxes
    plt.margins(x=0.2)
    
    # 3. Regression Plot
    plt.subplot(2, 2, 3)
    for i, market_type in enumerate(df['market_class'].unique()):
        subset = df[df['market_class'] == market_type]
        sns.regplot(x='risk_rating_score', y='volatility', data=subset,
                   label=market_type, scatter_kws={'alpha':0.3, 's':80},
                   color=sns.color_palette('viridis')[i],
                   line_kws={'lw':2})
    plt.title('ESG Score vs. Volatility Regression', pad=20, fontsize=14)
    plt.xlabel('ESG Score', fontsize=12)
    plt.ylabel('Volatility', fontsize=12)
    plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left')
    
    # 4. Grouped Violin Plot (alternative to KDE)
    plt.subplot(2, 2, 4)
    sns.violinplot(x='risk_rating_assessment', y='volatility', hue='market_class',
                  data=df, palette='viridis', split=True,
                  inner='quartile', scale='count', cut=0)
    
    plt.title('Volatility Distribution by Market and ESG Risk', pad=20, fontsize=14)
    plt.xlabel('ESG Risk Category', fontsize=12)
    plt.ylabel('Volatility', fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left')
    
    # Adjust layout
    plt.tight_layout(pad=3.0, h_pad=2.0, w_pad=2.0)
    plt.savefig('images/q3-esg_volatility_analysis.png', dpi=300, bbox_inches='tight')
    plt.show()

def analyze_volatility_relationships(df):
    """Statistical analysis of volatility relationships"""
    df = prepare_market_data(df)
    df_clean = prepare_volatility_data(df)
    
    # 1. Correlation analysis
    print("=== Correlation Analysis ===")
    corr_by_market = df_clean.groupby('market_class')[['risk_rating_score', 'volatility']].corr()
    print(corr_by_market)
    
    # 2. Regression models by market type
    print("\n=== Regression Results ===")
    results = {}
    for market in ['Developed', 'Emerging']:
        market_data = df_clean[df_clean['market_class'] == market]
        X = sm.add_constant(market_data['risk_rating_score'])
        model = sm.OLS(market_data['volatility'], X).fit()
        results[market] = model
        print(f"\n{market} Markets:")
        print(f"ESG Coefficient: {model.params[1]:.4f} (p={model.pvalues[1]:.4f})")
        print(f"R-squared: {model.rsquared:.4f}")
    
    # 3. ANOVA with interaction effect
    print("\n=== ANOVA with Interaction ===")
    try:
        model = ols('volatility ~ risk_rating_score * market_class', data=df_clean).fit()
        anova_table = sm.stats.anova_lm(model, typ=2)
        print(anova_table)
        
        # Check interaction significance
        interaction_p = anova_table.loc['risk_rating_score:market_class', 'PR(>F)']
        print(f"\nInteraction p-value: {interaction_p:.4f}")
        if interaction_p < 0.05:
            print("Significant interaction effect - ESG impact differs by market type")
        else:
            print("No significant interaction effect")
    except Exception as e:
        print(f"ANOVA failed: {str(e)}")
    
    # 4. Non-parametric tests
    print("\n=== Non-parametric Tests ===")
    for market in ['Developed', 'Emerging']:
        market_data = df_clean[df_clean['market_class'] == market]
        rho, pval = stats.spearmanr(market_data['risk_rating_score'], 
                                   market_data['volatility'])
        print(f"{market} Markets Spearman Correlation: rho={rho:.4f}, p={pval:.4f}")
    
    return df_clean, results

def interpret_results(model_results):
    """Provide plain English interpretation"""
    dev_coef = model_results['Developed'].params[1]
    emg_coef = model_results['Emerging'].params[1]
    
    print("\n=== Practical Interpretation ===")
    print(f"Developed Markets: Each 1-unit ESG improvement reduces volatility by {abs(dev_coef):.4f}")
    print(f"Emerging Markets: Each 1-unit ESG improvement reduces volatility by {abs(emg_coef):.4f}")
    
    difference = abs(dev_coef - emg_coef)
    if difference > 0.02:  # Threshold for practical significance
        print(f"\nSubstantial difference ({difference:.4f}) in ESG impact between markets")
        if abs(dev_coef) > abs(emg_coef):
            print("ESG scores more strongly influence volatility in Developed markets")
        else:
            print("ESG scores more strongly influence volatility in Emerging markets")
    else:
        print("\nMinimal practical difference in ESG impact between markets")

# Add to analysis pipeline:
# interpret_results(model_results)
```
